```python
import pandas as pd

movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

print("Movies Dataset Shape:", movies.shape)
print("Credits Dataset Shape:", credits.shape)
```

```python

credits.columns
```

```python
movies.columns
```

```python
movies = movies.merge(credits, on='title')

movies.shape
```

```python
movies = movies[['movie_id',
                 'title',
                 'overview',
                 'genres',
                 'keywords',
                 'cast',
                 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)

movies.shape
```

```python
import ast
```

```python
def convert(text):
    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)

movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(text):

    L = []

    counter = 0

    for i in ast.literal_eval(text):

        if counter != 3:

            L.append(i['name'])

            counter += 1

        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(text):

    L = []

    for i in ast.literal_eval(text):

        if i['job'] == 'Director':

            L.append(i['name'])

            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(
    lambda x:[i.replace(" ","") for i in x])

movies['keywords'] = movies['keywords'].apply(
    lambda x:[i.replace(" ","") for i in x])

movies['cast'] = movies['cast'].apply(
    lambda x:[i.replace(" ","") for i in x])

movies['crew'] = movies['crew'].apply(
    lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):

    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
movies['cast'] = movies['cast'].apply(
    lambda x:[i.replace(" ","") for i in x]
)
```

```python
def recommend(movie):

    matches = new_df[
        new_df['title'].str.contains(movie, case=False, na=False)
    ]

    if len(matches) == 0:
        print("Movie not found!")
        return

    movie_title = matches.iloc[0]['title']

    index = new_df[new_df['title'] == movie_title].index[0]

    distances = similarity[index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x:x[1]
    )

    print(f"\nRecommendations for {movie_title}:\n")

    for i in movie_list[1:11]:
        print(new_df.iloc[i[0]].title)
def movies_by_actor(actor_name):

    actor_name = actor_name.replace(" ", "").lower()

    result = movies[
        movies['cast'].apply(
            lambda cast:
            any(
                actor_name in actor.lower()
                for actor in cast
            )
        )
    ]

    return result[['title']]
```
